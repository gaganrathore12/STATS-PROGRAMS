```python
import pandas as pd
import numpy as np
```

```python
data = {
    'height':[150,155,160,165,170,175,180,185,190,195,200],
    'weight':[60,65,70,75,80,85,90,95,100,105,110]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.tail(2)
```

```python
karl_corr = df['height'].corr(df['weight'])
karl_corr
```

```python
spear_corr = df['height'].corr(df['weight'], method = 'spearman')
spear_corr
```

```python
df['rank1'] = df['height'].rank()
df['rank2'] = df['weight'].rank()
df.head()
```

```python
df['D']= df ['rank1']-df['rank2']
```

```python
df ['D2'] = df['D']**2
df
```

```python
# calculate mean and standard deviation od height and weight 
```

```python
mean2 = df['weight'].mean()
mean2
```

```python
df['x'] = df['height']-mean
df
```
